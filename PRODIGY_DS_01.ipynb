```python
#Missing values

data.isnull().any()
```

```python
# Histogram for Age in the dataset
# Continuous variable

plt.figure(figsize=(10,5))
plt.hist(data["Age"], bins=4,color='sandybrown',edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for Gender in the dataset
# Categorical variable

plt.figure(figsize = (10,5))
data['Gender'].value_counts().plot(kind='bar', color =['pink','lightblue'],edgecolor='black')
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for Ethnicity in the dataset
# Categorical variable

plt.figure(figsize = (10,5))
data['Ethnicity'].value_counts().plot(kind='bar', color =['salmon','lightgreen','yellow','cyan'],edgecolor='black')
plt.xlabel("Ethnicity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for Parental Education in the dataset
# Categorical variable

plt.figure(figsize = (10,5))
data['ParentalEducation'].value_counts().plot(kind='bar', color =['midnightblue','mediumblue','blue','royalblue','cornflowerblue'],edgecolor='black')
plt.xlabel("Parental Education")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for study time weekly of the students in the dataset
# Continuous variable

plt.figure(figsize=(20,10))
plt.hist(data["StudyTimeWeekly"], bins=100,color='mediumslateblue',edgecolor='black')
plt.xlabel("Study time weekly of the Students")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for Absences of the Students in the dataset
# Continuous variable

plt.figure(figsize=(10,5))
plt.hist(data["Absences"], bins=30,color='red',edgecolor='black')
plt.xlabel("Absences of the Students")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for Extracurricular Activities of the Students in the dataset
# Categorical variable

plt.figure(figsize = (10,5))
data['Extracurricular'].value_counts().plot(kind='bar', color =['orangered','limegreen'],edgecolor='black')
plt.xlabel("Extracurricular of the Students")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for GPA in the dataset
# Continuous variable

plt.figure(figsize=(10,5))
plt.hist(data["GPA"], bins=100,color='palegreen',edgecolor='black')
plt.xlabel("GPA of the Students")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for Grade Class in the dataset
# Categorical variable

plt.figure(figsize = (10,5))
data['GradeClass'].value_counts().plot(kind='bar', color =['darkgreen','seagreen','mediumseagreen','limegreen','lightgreen'],edgecolor='black')
plt.xlabel("Grade Class of the Students")
plt.ylabel("Frequency")
plt.show()
```
